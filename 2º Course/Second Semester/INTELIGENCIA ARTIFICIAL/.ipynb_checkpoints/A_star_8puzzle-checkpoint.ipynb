# Solución del 8 puzzle con A*

En esta práctica vamos a resolver el problema del 8 puzzle utilizando el algoritmo A*. Utilizaremos las funciones desarrolladas para la búsqueda en anchura. 
Para crear la heurística que guíe la búsqueda vamos definir la siguiente relajación:
En el problema del 8 puzzle solo se puede mover las fichas a la casilla que está vacía, es decir, las fichas no se pueden solapar. Podemos relajar el problema eliminando esta restricción, por lo tanto un problema relajado será resolver el problema del 8 puzzle en donde las fichas se pueden solapar. Como el coste de mover una ficha una casilla es 1, el coste de mover una ficha de su posición inicial a su posición final será la distancia de Manhattan entre las dos posiciones. Por lo tanto, el coste futuro de cada estado al estado final, será la suma de las distancias de Manhattan de la posición actual hasta la posición final de todas las fichas. Esta función será una heurística consitente y admisible para el problema del 8 puzzle original.

En la búsqueda A* los nodos deben guardar el coste y el valor de la heurística, por lo tanto creamos una clase nodo más completa:

```python
import copy 
class nodo:
    def __init__(self, estado, camino):
        self.estado = estado
        self.camino = camino
        self.coste = len(camino)
        self.heuristica = 0
        self.ch = self.coste + self.heuristica
    def set_heuristica(self, heuristica):
        self.heuristica = heuristica
        self.ch = self.coste + self.heuristica
    def actualiza_coste(self):
        self.coste = len(self.camino)
        
#INFORMACIÓN:
    #-El estado de los nodos de los algoritmos está representado por una lista de enteros, en vez de 
    #por una matriz (lista de listas). 
    #-Las acciones estan representadas por una lista de 4 enteros por los 4 posibles movimientos.
    #-El orden de los sucesores se crea en función de los siguientes movimientos:
        #1.Mover abajo = [1,0,0,0]
        #2.Mover arriba = [0,1,0,0]
        #3.Mover derecha = [0,0,1,0]
        #4.Mover izquierda = [0,0,0,1]
    #-Se recomienda que para visualizar el correcto funcionamiento del programa, se haga un run de todas las celdas 
    #a la vez debido a que puede haber problemas al ejecutar las gráficas por la interacción entre las clases con sus
    #atributos y métodos de los diferentes algoritmos (mismos nombres).
    #En la penúltima celda, se muestra el funcionamiento del algoritmo a estrella mostrando el camino, los nodos 
    #recorridos y el tiempo en ejecutar el algoritmo.
    #En la última celda, se realizan las gráficas
```

Copia en la siguiente celda las funciones de las acciones que creaste para la búsqueda en anchura, la función esfinal(), esposible() y la función aplica() (si las hiciste bien, seguirán funcionando con nodos de la nueva clase nodo) y  la lista de acciones. 

```python
def esfinal(nodo):
    final = [1,2,3,4,5,6,7,8,0]
    for i in range(len(nodo.estado)):
        if nodo.estado[i] !=  final[i]:
            return False
    return True

def esposible(accion, nodo):
    if nodo.estado[0] == 0 and (accion[3] == 1 or accion[1] == 1):
        return True
    elif nodo.estado[1] == 0 and (accion[0] == 0):
        return True
    elif nodo.estado[2] == 0 and (accion[2] == 1 or accion[1] == 1):
        return True
    elif nodo.estado[3] == 0 and (accion[2] == 0):
        return True
    elif nodo.estado[4] == 0:
        return True
    elif nodo.estado[5] == 0 and accion[3] == 0:
        return True
    elif nodo.estado[6] == 0 and accion[1] == 0 and accion[2] == 0:
        return True
    elif nodo.estado[7] == 0 and accion[1] == 0:
        return True
    elif nodo.estado[8] == 0 and accion[1] == 0 and accion[3] == 0:
        return True
    else:
        return False
    
def aplica(accion, nodo):
    if accion[0] == 1:
        sucesor = moverAbajo(nodo)
    elif accion[1] == 1:
        sucesor = moverArriba(nodo)
    elif accion[2] == 1:
        sucesor = moverDerecha(nodo)
    else:
        sucesor = moverIzquierda(nodo)
    return sucesor

import copy
def moverAbajo(nodo):
    nuevonodo = copy.deepcopy(nodo)
    accion = [1, 0, 0, 0]
    nuevonodo.camino.append(accion)
    
    for i in range(len(nuevonodo.estado)): ##guardas en la i la posición del 0
        if nuevonodo.estado[i] == 0:
            break
    aux = nuevonodo.estado[i-3]
    nuevonodo.estado[i-3] = nuevonodo.estado[i]
    nuevonodo.estado[i] = aux
    
    return nuevonodo
    
def moverArriba(nodo):
    nuevonodo = copy.deepcopy(nodo)
    accion = [0, 1, 0, 0]
    nuevonodo.camino.append(accion)
    
    for i in range(len(nuevonodo.estado)): ##guardas en la i la posición del 0
        if nuevonodo.estado[i] == 0:
            break
    aux = nuevonodo.estado[i+3]
    nuevonodo.estado[i+3] = nuevonodo.estado[i]
    nuevonodo.estado[i] = aux
    
    return nuevonodo
    
def moverIzquierda(nodo):
    nuevonodo = copy.deepcopy(nodo)
    accion = [0, 0, 0, 1]
    nuevonodo.camino.append(accion)
    
    for i in range(len(nuevonodo.estado)): ##guardas en la i la posición del 0
        if nuevonodo.estado[i] == 0:
            break
    aux = nuevonodo.estado[i+1]
    nuevonodo.estado[i+1] = nuevonodo.estado[i]
    nuevonodo.estado[i] = aux
    
    return nuevonodo
    
def moverDerecha(nodo):
    nuevonodo = copy.deepcopy(nodo)
    accion = [0, 0, 1, 0]
    nuevonodo.camino.append(accion)
    
    for i in range(len(nuevonodo.estado)): ##guardas en la i la posición del 0
        if nuevonodo.estado[i] == 0:
            break
    aux = nuevonodo.estado[i-1]
    nuevonodo.estado[i-1] = nuevonodo.estado[i]
    nuevonodo.estado[i] = aux
    
    return nuevonodo
    

def mostrarNodo(nodo):
    for i in range(0,len(nodo.estado),3):
        for j in range(i, i+3, 1):
            print(nodo.estado[j],end=" ")
            print(' ',end=" ")
        print()
def mostrarMatriz(matriz): 
    for i in range(3):
        for j in range(3):
            print(matriz[i][j],end=" ")
        print()
    print()
    
def argmin(abiertos):
    ind = 0
    for i in range(len(abiertos)):
        if abiertos[i].ch < abiertos[ind].ch:
            ind = i
    return ind
            
```

Modifica la función `estaen(lista, nodo)` para que devuelva **True** si existe **un nodo con el mismo estado y coste menor o igual** en la lista o **False** en caso contrario. 

```python
def estaen(lista, nodo):     
    for n in lista:
        if n.estado == nodo.estado and n.ch <= nodo.ch:
            return True       
    return False
```

Ahora debes crear la función heurística. La heurística es función del estado y es un valor que estima el coste hasta el estado final. Por tanto, la función que vamos a definir tendrá dos parámetros: el estado y el estado final:

```python
def heuristica(estado, est_final):
    value = 0
    for k in range (1, 9, 1):
        for i in range(len(estado)):
            if (estado[i] == k):
                break
        filaActual = (int)(i / 3)
        columnaActual = i % 3
        filaEstFinal = (int)((k - 1) / 3)
        columnaEstFinal = (int)((k - 1) % 3)
        value += abs(filaActual-filaEstFinal) + abs(columnaActual-columnaEstFinal)
    return value

estado=[8,7,6,5,4,3,2,1,0]
est_final=[1,2,3,4,5,6,7,8,0]
value = heuristica(estado,est_final)
```

Para comprobar en la práctica la complejidad del algoritmo vamos a comprobar el tiempo que tarda en ejecutarse y compararlo con la búsqueda en anchura. Ten en cuenta que después de generar los nodos sucesores tienes que calcular su heurística.

```python
from collections import deque
from IPython.display import display, clear_output
import time

def algoritmoEstrella (nodoinicial):

    start_time = time.time()

    # estado final:
    final = [[1,2,3],[4,5,6], [7,8,0]] 

    # nodo inicial
    #estado_inicial = [[1,2,3],[4,5,0], [7,8,6]] 
    #estado_inicial = [[1,2,0],[4,5,3], [7,8,6]] 
    #estado_inicial = [[1,0,2],[4,5,3], [7,8,6]]
    #estado_inicial = [[1,5,2],[4,0,3], [7,8,6]]
    #estado_inicial = [[1,5,2],[4,8,3], [7,0,6]]
    #estado_inicial = [[1,5,2],[4,8,3], [0,7,6]]
    #estado_inicial = [[1,5,2],[0,8,3], [4,7,6]]
    #estado_inicial = [[0,5,2],[1,8,3], [4,7,6]]
    #estado_inicial = [[5,0,2],[1,8,3], [4,7,6]]
    #estado_inicial = [[5,2,0],[1,8,3], [4,7,6]]
    #estado_inicial = [[5,2,3],[1,8,0], [4,7,6]]
    #estado_inicial= [[5,2,3],[1,0,8], [4,7,6]]
    #estado_inicial = [[5,2,3],[1,7,8], [4,0,6]]
    #estado_inicial = [[5,2,3],[1,7,8], [0,4,6]]

    # Busqueda A*

    print("NODO INICIAL: \n")
    mostrarNodo(nodoinicial)
    accion1 = [1,0,0,0]
    accion2 = [0,1,0,0]
    accion3 = [0,0,1,0]
    accion4 = [0,0,0,1]
    listaacciones = [accion1, accion2, accion3, accion4]
    abiertos = []
    abiertos.append(nodoinicial)
    cerrados=[]
    sucesores=[]
    nuevossucesores=[]
    contador = 0
    while len(abiertos) > 0:
        IndMin = argmin(abiertos)
        actual = abiertos [IndMin]
        del abiertos[IndMin]
        cerrados.append(actual)
        contador = contador + 1
        if esfinal(actual):
            seconds = time.time() - start_time
            print("\nNODO FINAL:\n")
            mostrarNodo(actual)
            print()
            for i in range(len(actual.camino)):
                if actual.camino[i] == [1,0,0,0]:
                    print("Abajo")
                elif actual.camino[i] == [0,1,0,0]:
                    print("Arriba")
                elif actual.camino[i] == [0,0,1,0]:
                    print("Derecha")
                else:
                    print("Izquierda")
            print()
            print(seconds)
            print('\nTime Taken:', time.strftime("%H:%M:%S",time.gmtime(seconds)))
            return contador
        else:
            sucesores.clear()
            nuevossucesores.clear()

            for act in listaacciones:
                if esposible(act,actual):
                    sucesor = aplica(act,actual)
                    sucesor.actualiza_coste()
                    heurist = heuristica(sucesor.estado,final)
                    sucesor.set_heuristica(heurist)
                    sucesores.append(sucesor)
            for i in range(len(sucesores)):
                if estaen(abiertos,sucesores[i]) == False and estaen(cerrados,sucesores[i]) == False:
                    nuevossucesores.append(sucesores[i])
            abiertos += (nuevossucesores) 
        ##mostrarNodo(actual)##muestras todos los nodos que coges con abiertos[0] hasta que llegas al estado final
    return -1
nodoinicial = nodo([5,2,3,1,7,8,0,4,6], [])
print("Nodos actuales: ", algoritmoEstrella(nodoinicial))
```

```python
import matplotlib.pyplot as plt
import random 

from collections import deque
from IPython.display import display, clear_output
import time

def algoritmoEstrella (nodoinicial):

    start_time = time.time()

    # estado final:
    final = [[1,2,3],[4,5,6], [7,8,0]] 

    # Busqueda A*
    accion1 = [1,0,0,0]
    accion2 = [0,1,0,0]
    accion3 = [0,0,1,0]
    accion4 = [0,0,0,1]
    listaacciones = [accion1, accion2, accion3, accion4]
    abiertos = []
    abiertos.append(nodoinicial)
    cerrados=[]
    sucesores=[]
    nuevossucesores=[]
    contador = 0
    while len(abiertos) > 0:
        IndMin = argmin(abiertos)
        actual = abiertos [IndMin]
        del abiertos[IndMin]
        cerrados.append(actual)
        contador = contador + 1
        if esfinal(actual):
            seconds = time.time() - start_time
            return contador
        else:
            sucesores.clear()
            nuevossucesores.clear()

            for act in listaacciones:
                if esposible(act,actual):
                    sucesor = aplica(act,actual)
                    sucesor.actualiza_coste()
                    heurist = heuristica(sucesor.estado,final)
                    sucesor.set_heuristica(heurist)
                    sucesores.append(sucesor)
            for i in range(len(sucesores)):
                if estaen(abiertos,sucesores[i]) == False and estaen(cerrados,sucesores[i]) == False:
                    nuevossucesores.append(sucesores[i])
            abiertos += (nuevossucesores) 
        ##mostrarNodo(actual)##muestras todos los nodos que coges con abiertos[0] hasta que llegas al estado final
    return -1
estado_inicial = [[1,2,3,4,5,0,7,8,6], [1,2,0,4,5,3,7,8,6], [1,0,2,4,5,3,7,8,6], 
                  [1,5,2,4,0,3,7,8,6], [1,5,2,4,8,3,7,0,6], [1,5,2,4,8,3,0,7,6], [1,5,2,0,8,3,4,7,6], 
                  [0,5,2,1,8,3,4,7,6], [5,0,2,1,8,3,4,7,6], [5,2,0,1,8,3,4,7,6], [5,2,3,1,8,0,4,7,6],
                  [5,2,3,1,0,8,4,7,6], [5,2,3,1,7,8,4,0,6], [5,2,3,1,7,8,0,4,6]]
plt.subplot(1,2,1)
x1 = range(1,14)
y1 = [algoritmoEstrella(nodo(estado_inicial[i],[])) for i in range(len(estado_inicial)-1)] 
plt.plot(x1, y1, 'go--', linewidth=2, markersize=10) 
plt.title("Algoritmo A*")
plt.suptitle("Nodos recorridos")

class nodo:
    def __init__(self, estado, camino):
        self.estado = estado
        self.camino = camino
def estaen(lista, nodo):     
    for n in lista:
        if n.estado == nodo.estado:
            return True       
    return False
def esfinal(nodo):
    final = [1,2,3,4,5,6,7,8,0]
    for i in range(len(nodo.estado)):
        if nodo.estado[i] !=  final[i]:
            return False
    return True
def moverArriba(nodo):
    nuevonodo = copy.deepcopy(nodo)
    accion = [1, 0, 0, 0]
    nuevonodo.camino.append(accion)

    for i in range(len(nuevonodo.estado)): ##guardas en la i la posición del 0
        if nuevonodo.estado[i] == 0:
            break
    aux = nuevonodo.estado[i-3]
    nuevonodo.estado[i-3] = nuevonodo.estado[i]
    nuevonodo.estado[i] = aux

    return nuevonodo

def moverAbajo(nodo):
    nuevonodo = copy.deepcopy(nodo)
    accion = [0, 1, 0, 0]
    nuevonodo.camino.append(accion)

    for i in range(len(nuevonodo.estado)): ##guardas en la i la posición del 0
        if nuevonodo.estado[i] == 0:
            break
    aux = nuevonodo.estado[i+3]
    nuevonodo.estado[i+3] = nuevonodo.estado[i]
    nuevonodo.estado[i] = aux

    return nuevonodo

def moverDerecha(nodo):
    nuevonodo = copy.deepcopy(nodo)
    accion = [0, 0, 0, 1]
    nuevonodo.camino.append(accion)

    for i in range(len(nuevonodo.estado)): ##guardas en la i la posición del 0
        if nuevonodo.estado[i] == 0:
            break
    aux = nuevonodo.estado[i+1]
    nuevonodo.estado[i+1] = nuevonodo.estado[i]
    nuevonodo.estado[i] = aux

    return nuevonodo

def moverIzquierda(nodo):
    nuevonodo = copy.deepcopy(nodo)
    accion = [0, 0, 1, 0]
    nuevonodo.camino.append(accion)

    for i in range(len(nuevonodo.estado)): ##guardas en la i la posición del 0
        if nuevonodo.estado[i] == 0:
            break
    aux = nuevonodo.estado[i-1]
    nuevonodo.estado[i-1] = nuevonodo.estado[i]
    nuevonodo.estado[i] = aux

    return nuevonodo
def aplica(accion, nodo):
    if accion[0] == 1:
        sucesor = moverArriba(nodo)
    elif accion[1] == 1:
        sucesor = moverAbajo(nodo)
    elif accion[2] == 1:
        sucesor = moverIzquierda(nodo)
    else:
        sucesor = moverDerecha(nodo)
    return sucesor
def esposible(accion, nodo):
    if nodo.estado[0] == 0 and (accion[3] == 1 or accion[1] == 1):
        return True
    elif nodo.estado[1] == 0 and (accion[0] == 0):
        return True
    elif nodo.estado[2] == 0 and (accion[2] == 1 or accion[1] == 1):
        return True
    elif nodo.estado[3] == 0 and (accion[2] == 0):
        return True
    elif nodo.estado[4] == 0:
        return True
    elif nodo.estado[5] == 0 and accion[3] == 0:
        return True
    elif nodo.estado[6] == 0 and accion[1] == 0 and accion[2] == 0:
        return True
    elif nodo.estado[7] == 0 and accion[1] == 0:
        return True
    elif nodo.estado[8] == 0 and accion[1] == 0 and accion[3] == 0:
        return True
    else:
        return False
    
def algoritmoAnchura (nodoinicial):
    start_time = time.time()
    accion1 = [1,0,0,0]
    accion2 = [0,1,0,0]
    accion3 = [0,0,1,0]
    accion4 = [0,0,0,1]
    listaacciones = [accion1, accion2, accion3, accion4]
    abiertos = []
    abiertos.append(nodoinicial)
    cerrados=[]
    sucesores=[]
    nuevossucesores=[]
    contador = 0
    while len(abiertos) > 0:
        actual = abiertos[0]
        del abiertos[0]
        cerrados.append(actual)
        contador = contador + 1
        if esfinal(actual):
            seconds = time.time() - start_time
            return contador
        else:
            sucesores.clear()
            nuevossucesores.clear()
            for act in listaacciones:
                if esposible(act,actual):
                    sucesores.append(aplica(act,actual))
            for i in range(len(sucesores)):
                if estaen(abiertos,sucesores[i]) == False and estaen(cerrados,sucesores[i]) == False:
                    nuevossucesores.append(sucesores[i])
            abiertos += (nuevossucesores) 
    return -1
plt.subplot(1,2,2)
x2 = range(1,14)
y2 = [algoritmoAnchura(nodo(estado_inicial[i],[])) for i in range(len(estado_inicial)-1)] 
plt.plot(x2, y2, 'go--', linewidth=2, markersize=10) 
plt.title("Algoritmo Anchura")
plt.suptitle("Nodos recorridos")
```
