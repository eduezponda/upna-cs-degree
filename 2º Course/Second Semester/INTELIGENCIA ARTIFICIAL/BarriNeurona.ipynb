```python
import pandas as pd
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import numpy as np

datos = pd.read_csv("heart_2020.csv")
datos = datos.drop(['Unnamed: 0'], axis=1)
datos = shuffle(datos)
datos.head()
```

```python
datos.describe()
```

```python
from sklearn.model_selection import train_test_split
# Ten en cuenta que la variable y tiene valores (0 y 1)
y = np.array(datos['HeartDisease'])
datos = datos.drop(['HeartDisease'], axis=1)
X = np.array(datos)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state= 2)

for i in range(len(y_train)):
    if(y_train[i] == 0):
        y_train[i] = -1
for i in range(len(y_test)):
    if(y_test[i] == 0):
        y_test[i] = -1

#Comprobamos que los valores sean -1 o 1
print(y_train, y_test)
print(X_test)
```

```python
import random
def perceptron(w,x):
    c = np.dot(w,x)
    if c > 0:
        return 1;
    elif c < 0:
        return -1
    return 0
    
```

```python
def margen(x, y, w):
    s = np.dot(w,x) * y
    return s

def perdidaBisagra(x,y,w):
    sum = 0
    for i in range(len(x)):
        sum = sum + max(1-margen(x[i], y[i], w),0)
    sum = sum / len(x)
    return sum

def coste(X, y, w):
    loss = 0
    for i in range(len(X)):
        c = np.dot(w,X[i])
        loss += (y[i]-c)**2
    return loss/(len(X))
```

```python
def gradientes(x,y,w):
    grad = []
    for i in range(len(w)):
        sum = 0
        for j in range(len(x)):
            valor = 0
            if(margen(x[j], y[j], w) < 1):
                valor = -x[j][i]*y[j]
            else:
                valor = 0
            sum = sum + valor
        grad.append(sum/len(x))
    return grad

def descensoGradiente(x, y, w, alpha, iters):
    loss = []
    for i in range(iters):
        g = gradientes(x,y,w)
        for j in range(len(g)):
            w[j] = w[j] - alpha*g[j]
        loss.append(perdidaBisagra(x,y,w))
    return loss, w
    
    
    
w_aux = np.random.rand(len(X_train[0]))
l, w = descensoGradiente(X_train, y_train, w_aux, 0.01, 100)
plt.plot(l) 
plt.show()
```

Podemos observar que el error diminuye.

```python
def prediccion(X, w):
    y_p = []
    for i in range(len(X)):
        y_p.append(perceptron(w, X[i]))
    return y_p

def matriz_confusion(y, y_p):
    vp = 0
    vn = 0
    fn = 0
    fp = 0
    
    for i in range(len(y)):
        if(y_p[i] == 1):
            if(y[i] == 1):
                vp +=1
            elif(y[i] == -1):
                fp += 1
        elif(y_p[i] == -1):
            if(y[i] == 1):
                fn +=1
            elif(y[i] == -1):
                vn += 1
                
    m = []
    m.append([vp,fp])
    m.append([fn,vn])
    return m

y_p = prediccion(X_test, w)
m = matriz_confusion(y_test, y_p)
print(m)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm1 = confusion_matrix(y_test, y_p)
print(cm1)
```
