## Proyecto Redes Neuronales. Compración Perceptrón simple vs Perceptrón multicapa en  clasificación

Según los CDC, las enfermedades del corazón son una de las principales causas de muerte. Aproximadamente la mitad de los estadounidenses (47%) presentan al menos uno de los tres factores de riesgo clave de las enfermedades cardíacas: presión arterial alta, colesterol alto y tabaquismo. Otros indicadores clave son la condición de diabético, la obesidad (IMC elevado), la falta de actividad física o el consumo excesivo de alcohol. Detectar y prevenir los factores que más influyen en las enfermedades del corazón es muy importante en la asistencia sanitaria. Los avances informáticos, por su parte, permiten aplicar métodos de aprendizaje automático para detectar "patrones" a partir de los datos que puedan predecir el estado de un paciente.

En este proyecto vas a **comparar la eficacia de dos métodos diferentes en la predicción del riesgo de padecer una enfermedad cardiaca**. 

El conjunto de datos procede de los CDC y es una parte importante del Sistema de Vigilancia de los Factores de Riesgo en el Comportamiento, que realiza encuestas telefónicas anuales para recopilar datos sobre el estado de salud de los residentes en EE.UU. En la siguiente celda se leen los datos y se generan los conjuntos de entrenamiento y de test.

```python
import pandas as pd
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import numpy as np

datos = pd.read_csv("heart_2020.csv")
datos = datos.drop(['Unnamed: 0'], axis=1)
datos = shuffle(datos)
datos.head()
```

```python
datos.describe()
```

```python
from sklearn.model_selection import train_test_split
# Ten en cuenta que la variable y tiene valores (0 y 1)
y = np.array(datos['HeartDisease'])
datos = datos.drop(['HeartDisease'], axis=1)
X = np.array(datos)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state= 2)

y_train[y_train == 0] = -1
y_test[y_test == 0] = -1
 
print(y_train, y_test)
```

```python
def perceptron(x,w):
    return np.sign(np.dot(x,w))
def coste(X, y, w):
    loss = 0
    for i in range(len(X)):
        loss += max(1-(np.dot(w,X[i])*y[i]),0)        
    return loss/len(X)
def gradientes(X,y,w):
    grad = []
    for i in range(len(w)):
        sum = 0
        for j in range(len(X)):
            valor = 0
            if(np.dot(X[j],w) * y[j] < 1):
                valor = -X[j][i]*y[j]
            else:
                valor = 0
            sum = sum + valor
        grad.append(sum/len(X))
    return grad
def descensoGradiente(X, y, w, alpha = 0.01, iters = 100):
    loss = []
    for i in range(iters):
        grad = gradientes(X,y,w)
        for j in range(len(grad)):
            w[j] -= alpha*grad[j]
        loss.append(coste(X,y,w))
    return loss, w
```

```python
w_ini = np.random.rand(len(X_train[0]))
l, w = descensoGradiente(X_train, y_train, w_ini)
plt.plot(l) 
plt.show()
```

La librería sklearn tiene implementada una clase MLPClassifier que es un perceptrón multicapa con el algoritmo de backpropagation que podemos utilizar en la siguiente celda se crea un preceptrón con una capa oculta y dos neuronas para que se ajuste a la función XOR.
Para ayuda sobre la clase y sus parámetros visitar: https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html

```python
X = np.array([[2,2], [0,0], [2,0], [0,2]])
y = np.array([1, 1, 0, 0])
```

```python
from sklearn.neural_network import MLPClassifier
# Función para crear el perceptrón multicapa
pmc = MLPClassifier(solver='lbfgs', alpha=1e-2, hidden_layer_sizes=(2,), activation='logistic', random_state=1)
# Al llamar a fit se realiza el entrenamiento
pmc.fit(X, y)

y_p = pmc.predict(X)

#Valores predichos
print(y_p)
# Valores de los pesos de la red
print(pmc.coefs_)
# Valores de los pesos correspondiente a los bias (x_0)
print(pmc.intercepts_)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y, y_p)
print(cm)
```

```python
def predict_y(X, w):
    return [perceptron(w, x) for x in X]

y_pred = predict_y(X_test, w)
matriz1 = confusion_matrix(y_test, y_pred)
print(matriz1)

vp = matriz1[0][0]
vn = matriz1[1][1]
fp = matriz1[1][0]
fn = matriz1[0][1]
    
accuracy = (vp + vn) / (vp + vn + fp + fn)
recall = vp / (vp + fn)
precision = vp / (vp + fp)
tnr = vn / (vn + fp)
```

```python
#Red neuronal con una capa con tres neuronas
pmc = MLPClassifier(solver='lbfgs', alpha=1e-2, hidden_layer_sizes=(3,), activation='relu', random_state=1,max_iter=1000)
pmc.fit(X_train, y_train)
y_p = pmc.predict(X_test)
matriz2 = confusion_matrix(y_test, y_p)
print(matriz2)

#Red neuronal con dos capas, cada una de ellas con dos neuronas
pmc2 = MLPClassifier(solver='lbfgs', alpha=.1e-2, hidden_layer_sizes=(2,2), activation='logistic', random_state=1,max_iter=1000)
pmc2.fit(X_train, y_train)
y_p = pmc2.predict(X_test)
matriz3 = confusion_matrix(y_test, y_p)
print(matriz3)
```

```python
def accuracy(m):
    vn = m[0][0]
    vp = m[1][1]
    fp = m[0][1]
    fn = m[1][0]
    return (vp +vn) / (vp + fp + vn + fn)

def recall(m):
    vn = m[0][0]
    vp = m[1][1]
    fp = m[0][1]
    fn = m[1][0]
    return (vp) / (vp + fn)

def precision(m):
    vn = m[0][0]
    vp = m[1][1]
    fp = m[0][1]
    fn = m[1][0]
    return (vp) / (vp + fp)

def tnr(m):
    vn = m[0][0]
    vp = m[1][1]
    fp = m[0][1]
    fn = m[1][0]
    return (vn) / (vn + fp)

print("Acurracy perceptrón simple: ", accuracy(matriz1))
print("Accuracy primera neurona multicapa: ", accuracy(matriz2))
print("Accuracy segunda neurona multicapa: ", accuracy(matriz1))
print()
print("Recall perceptrón simple: ", recall(matriz1))
print("Recall primera neurona multicapa: ", recall(matriz2))
print("Recall segunda neurona multicapa: ", recall(matriz3))
print()
print("Precision perceptrón simple: ", precision(matriz1))
print("Precision primera neurona multicapa: ", precision(matriz2))
print("Precision segunda neurona multicapa: ", precision(matriz3))
print()
print("TNR perceptrón simple: ", tnr(matriz1))
print("TNR primera neurona multicapa: ", tnr(matriz2))
print("TNR segunda neurona multicapa: ", tnr(matriz3))
```

Como conclusión, al querer comparar la eficacia de dos métodos diferentes en la predicción del riesgo de padecer una enfermedad cardiaca, tendremos que observar detenidamente el recall. Nuestro objetivo es maximizar el recall para que no haya falsos negativos (fn). No queremos falsos negativos debido a que al poder padecer una enfermedad, lo más importante no es errar cuando el paciente no tiene ninguna enfermedad, sino cuando puede padecerla no detectarla. 

Como se puede observar, la segunda neurona multicapa tiene el mayor recall sin perder precisión a la vez. Por lo tanto, la segunda neurona multicapa será la más eficiente.
