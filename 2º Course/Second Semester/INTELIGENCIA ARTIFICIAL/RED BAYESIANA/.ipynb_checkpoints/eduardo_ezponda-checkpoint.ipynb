***EXPLICACI√ìN DEL PROBLEMA:***

En el siguiente proyecto se crea una red bayesiana acerca de la probabilidad de que una persona padezca diabetes,
teniendo en cuenta una serie de factores. La diabetes es una enfermedad metab√≥lica cr√≥nica caracterizada por niveles 
elevados de glucosa en sangre (o az√∫car en sangre), que con el tiempo conduce a da√±os graves en el coraz√≥n, 
los vasos sangu√≠neos, los ojos, los ri√±ones y los nervios.
Se estima que la diabetes afecta al 9,3% de la poblaci√≥n mundial. Por lo tanto, es una de enfermedades m√°s comunes.

***OBJETIVO:***

   -A partir de unas espec√≠ficas variables y sus respectivas probabilidades,
    se determinar√° la probabilidad de padecer diabetes.

***EXPLICACI√ìN DE LAS VARIABLES:***

Se han utilizado 13 variables:

- G√©nero: ¬øEs hombre o mujer?: Hombre (0) ; Mujer (1). VARIABLE DE OBSERVACI√ìN

- Tabaco: ¬øConsume tabaco: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR

- Az√∫car: ¬øConsume m√°s azucar de lo recomendado?: Verdadero (0); Falso (1).FACTORES DE TIPO PROM.

- Deporte: ¬øRealiza deporte?: Verdadero (0) ; Falso (1). FACTORES DE TIPO INHIBIDOR

- Estr√©s: ¬øPadece de estres continuado?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR

- Alcohol: ¬øConsume alcohol regularmente?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR

- Sue√±o: ¬øPadece falta de sue√±o?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR

- Drogas: ¬øConsume drogas?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR

- Colesterol: ¬øTiene colesterol alto?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR

- Presi√≥n Arterial: ¬øTiene la presi√≥n arterial alta?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR

- Sobrepeso: ¬øPadece sobrepeso?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR

- Salud mental: ¬øSe encuentra en un mal estado psicol√≥gico?: Verdadero (0) ; Falso (1). FACTORES INHIBIDOR

- Diabetes: VARIABLE DE OBJETIVO
    
Todas las variables son de tipo booleano (representan cantidades num√©ricas).
En el caso de que se cumpla dicho factor de la variable, el valor ser√° 0. En caso contrario, su respectivo valor
ser√° 1. 

Variables cualitativas(Dicot√≥micas): G√©nero
Variables cuantitativas(Discretas): Tabaco, Sobrepeso, Presi√≥n Arterial, Az√∫car, Deporte, Drogas, Estr√©s, Alcohol, Sue√±o, Salud Mental y Colesterol

***ALGORITMO DE PEARL:***

[G√©nero, Tabaco, Deporte, Colesterol, Sue√±o, Estr√©s, Az√∫car, Alcohol, Presi√≥n Arterial, Drogas, Diabetes, Salud mental, Sobrepeso]

A√±adimos la primera variable G√©nero. A continuaci√≥n, las variables Tabaco y Deporte son dependientes de G√©nero, 
ya que dependiendo de si el sujeto es hombre o mujer, tendr√° m√°s o menos probabilidad de fumar o hacer deporte.
Todas las variables dependientes tendr√°n que ser contrarrestadas con un arco adicional que las una.
Colesterol, Sue√±o, Estr√©s, Az√∫car y Alcohol no tienen ninguna dependencia ni son consecuencia de ninguna variable. Por lo tanto, ser√°n ra√≠ces y no tendr√°n padres. La presi√≥n arterial depende del az√∫car y del nivel de colesterol que tenga el sujeto, por lo que se unir√°n con un arco. El consumo de drogas es dependiente del alcohol y del tabaco, por lo que estar√°n unidos. 
En funci√≥n del nivel de la presi√≥n arterial del sujeto, y de si realiza deporte regularmente, el sujeto tendr√° sobrepeso. Por √∫ltimo, la salud mental es dependiente de tener estr√©s continuado y falta de sue√±o. Por lo tanto, estar√° contrarrestado con un arco adicional.

Finalmente, la variable Diabetes ser√° la dependiente de todas las anteriores variables ya que ser√°n las causas de 
que el sujeto padezca o no diabetes. Con lo cual, se a√±adir√°n arcos que unir√°n cada variable con Diabetes.

```python
#CLASE NODO Y FUNCIONES
import random
class nodo:
    def __init__(self, padres, hijos, p):
        self.padres = padres
        self.hijos = hijos
        self.p = p
        
        self.num_states = 2
        self.states = [0, 1]
        self.state = None
        
    def sample(self, estado_padres):
        prob = random.random()
        if (len(estado_padres) == 0):
            if (prob < self.p[0]):
                self.state = 0
            else:
                self.state = 1
        elif (len(estado_padres) == 1):
            if prob < self.p[estado_padres[0]][0]:
                self.state = 0
            else:
                self.state = 1
        elif (len(estado_padres) == 2):
            if prob < self.p[estado_padres[0]][estado_padres[1]][0]:
                self.state = 0
            else:
                self.state = 1
        else:
            if prob < self.p[estado_padres[0]][estado_padres[1]][estado_padres[2]][0]:
                self.state = 0
            else:
                self.state = 1
                  
#VARIABLES RED BAYESIANA:
#G√©nero
pG = [0.505, 0.495]
Genero = nodo([], [1,2], pG)
#Tabaco
pT = [[0.16,0.84],[0.23,0.77]]
Tabaco = nodo([0],[4],pT)
#Deporte
pD = [[0.65,0.35],[0.54,0.46]]
Deporte = nodo([0],[8],pD)
#Alcohol
pA = [0.63,0.37]
Alcohol = nodo([],[4],pA)
#Drogas
pD = [[[0.9,0.1],[0.4,0.6]],[[0.25,0.75],[0.05,0.95]]]
Drogas = nodo([1,3],[12],pD)
#Azucar
pAz = [0.54,0.46]
Azucar = nodo([],[7],pAz)
#Colesterol
pC = [0.55,0.45]
Colesterol = nodo([],[7],pC)
#Presion_Arterial
pP = [[[0.8,0.2],[0.3,0.7]],[[0.4,0.6],[0.2,0.8]]]
Presion_Arterial = nodo([5,6],[8],pP)
#Sobrepeso
pS = [[[0.15,0.85],[0.05,0.95]],[[0.8,0.2],[0.5,0.5]]]
Sobrepeso = nodo([2,7],[12],pS)
#Sue√±o
pSu = [0.4,0.6]
Sue√±o = nodo([],[11],pSu)
#Estres
pE = [0.42,0.58]
Estres = nodo([],[11],pE)
#Salud_Mental
pSM = [[[0.95,0.05],[0.7,0.3]],[[0.3,0.7],[0.15,0.85]]]
Salud_Mental = nodo([9,10],[12],pSM)
#Diabetes
pDia = [[[[0.75,0.25],[0.7,0.3]],[[0.35,0.65],[0.3,0.7]]],[[[0.45,0.55],[0.4,0.6]],[[0.05,0.95],[0.01,0.99]] ] ]
Diabetes = nodo([4,8,11],[],pDia)

Grafo = {0:Genero, 1:Tabaco, 2:Deporte, 3:Alcohol, 4:Drogas, 5:Azucar, 6:Colesterol, 7:Presion_Arterial,
8:Sobrepeso, 9:Sue√±o, 10:Estres , 11:Salud_Mental, 12:Diabetes}

raices = [Genero,Azucar,Colesterol,Sue√±o,Estres,Alcohol]

def imprime(g):
    totalstate={}
    for nodo in g.keys():
        totalstate[nodo] = g[nodo].state
    print(totalstate)
imprime(Grafo)
def reset_grafo(g):
    for nodo in g.keys():
        g[nodo].state = None
reset_grafo(Grafo)
imprime(Grafo)
from collections import deque
def recorrido_anchura(node, graph):
    abiertos = deque()
    abiertos.extend(node)
    cerrados = []
    actual = None
    while (len(abiertos) > 0):
        actual = abiertos.popleft()
        cerrados.append(actual)
        estado_padres =  []
        for i in range(len(actual.padres)):
            estado_padres.append(graph[actual.padres[i]].state)
        actual.sample(estado_padres)
        sucesores = actual.hijos
        nuevos_sucesores = []
        for sucesor in sucesores:
            sucesor = graph[sucesor]
            if not sucesor in abiertos and not sucesor in cerrados:
                nuevos_sucesores.append(sucesor)
        abiertos.extend(nuevos_sucesores)
    return graph
reset_grafo(Grafo)
imprime(recorrido_anchura(raices, Grafo))
def sample_grafo(node, graph):
    reset_grafo(graph)
    states = recorrido_anchura(node,graph)
    return states
imprime(sample_grafo(raices,Grafo))
def match (graph, predicate):
    for keys in predicate:
        if graph[keys].state != predicate[keys]:
            return False
    return True

def infer(query, evidence, graph, num = 20000):
    n_evidence = 0
    n_query_and_evidence = 0
    for i in range(num):
        sample = sample_grafo(raices,graph)
        if (match(sample,evidence)):
            n_evidence += 1
            if (match(sample,query)):
                n_query_and_evidence += 1
    return n_query_and_evidence / n_evidence
```

***OBSERVACI√ìN DE RESULTADOS:***

Adem√°s, aquellas personas que realicen deporte, normalmente no tomar√°n drogas ni tendr√°n sobrepeso, por lo que la probabilidad de tener diabetes ser√° peque√±a.
Finalmente, estar en un buen o en un mal estado psicol√≥gico no variar√° en gran medida la probabilidad final del sujeto, debido a que el estado f√≠sico y los buenos h√°bitos tendr√°n mayor importancia.

- Las ventajas de la red proporcionada son principalmente que se encuentran como variables las causas de mayor importancia de padecer la enfermedad de la diabetes. Por lo tanto, todas aquellas combinaciones de la red establecer√°n una probabilidad aproximada a la que se puede asemejar en la realidad.

- Las desventajas de la red son que existen m√∫ltiples causas sobre la enfermedad que no se representan en la red como la etnia o la edad. No se encuentran debido a que la extensi√≥n de variables ser√≠a muy grande y la cantidad de variables y de probabilidades har√≠a compleja la estructura de la red. Por √∫ltimo, otro posible inconveniente ser√≠a que parte de las probabilidades son aproximadas y no tienen una fuente que las verifique debido a la dificultad del conocimiento y de la b√∫squeda de √©stas.
