***DIABETES:***

***EXPLICACIÓN DEL PROBLEMA:***

En el siguiente proyecto se crea una red bayesiana acerca de la probabilidad de que una persona padezca diabetes,
teniendo en cuenta una serie de factores. La diabetes es una enfermedad metabólica crónica caracterizada por niveles 
elevados de glucosa en sangre (o azúcar en sangre), que con el tiempo conduce a daños graves en el corazón, 
los vasos sanguíneos, los ojos, los riñones y los nervios.
Se estima que la diabetes afecta al 9,3% de la población mundial. Por lo tanto, es una de enfermedades más comunes.

***OBJETIVO:***

   -A partir de unas específicas variables y sus respectivas probabilidades,
    se determinará la probabilidad de padecer diabetes.

***EXPLICACIÓN DE LAS VARIABLES:***

Se han utilizado 13 variables:

    -Género: ¿Es hombre o mujer?: Hombre (0) ; Mujer (1). VARIABLE DE OBSERVACIÓN
    
    -Tabaco: ¿Consume tabaco: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR
    
    -Azúcar: ¿Consume más azucar de lo recomendado?: Verdadero (0); Falso (1).FACTORES DE TIPO PROM.
    
    -Deporte: ¿Realiza deporte?: Verdadero (0) ; Falso (1). FACTORES DE TIPO INHIBIDOR
    
    -Estrés: ¿Padece de estres continuado?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR
    
    -Alcohol: ¿Consume alcohol regularmente?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR
    
    -Sueño: ¿Padece falta de sueño?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR
    
    -Drogas: ¿Consume drogas?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR
    
    -Colesterol: ¿Tiene colesterol alto?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR
    
    -Presión Arterial: ¿Tiene la presión arterial alta?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR
    
    -Sobrepeso: ¿Padece sobrepeso?: Verdadero (0) ; Falso (1). FACTORES DE TIPO PROMOTOR
    
    -Salud mental: ¿Se encuentra en un mal estado psicológico?: Verdadero (0) ; Falso (1). FACTORES INHIBIDOR
    
    -Diabetes: VARIABLE DE OBJETIVO
    
Todas las variables son de tipo booleano (representan cantidades numéricas).
En el caso de que se cumpla dicho factor de la variable, el valor será 0. En caso contrario, su respectivo valor
será 1. 

Variables cualitativas(Dicotómicas): Género
Variables cuantitativas(Discretas): Tabaco, Sobrepeso, Presión Arterial, Azúcar, Deporte, Drogas, Estrés, Alcohol, Sueño, Salud Mental y Colesterol

***ALGORITMO DE PEARL:***
[Género, Tabaco, Deporte, Colesterol, Sueño, Estrés, Azúcar, Alcohol, Presión Arterial, Drogas, Diabetes, Salud mental, Sobrepeso]

Añadimos la primera variable Género. A continuación, las variables Tabaco y Deporte son dependientes de Género, 
ya que dependiendo de si el sujeto es hombre o mujer, tendrá más o menos probabilidad de fumar o hacer deporte.
Todas las variables dependientes tendrán que ser contrarrestadas con un arco adicional que las una.
Colesterol, Sueño, Estrés, Azúcar y Alcohol no tienen ninguna dependencia ni son consecuencia de ninguna variable. Por lo tanto, serán raíces y no tendrán padres. La presión arterial depende del azúcar y del nivel de colesterol que tenga el sujeto, por lo que se unirán con un arco. El consumo de drogas es dependiente del alcohol y del tabaco, por lo que estarán unidos. 
En función del nivel de la presión arterial del sujeto, y de si realiza deporte regularmente, el sujeto tendrá sobrepeso. Por último, la salud mental es dependiente de tener estrés continuado y falta de sueño. Por lo tanto, estará contrarrestado con un arco adicional.

Finalmente, la variable Diabetes será la dependiente de todas las anteriores variables ya que serán las causas de 
que el sujeto padezca o no diabetes. Con lo cual, se añadirán arcos que unirán cada variable con Diabetes.
