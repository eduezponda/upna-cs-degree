# Bisection method

Write a program that, given an interval containing the root of a trascedent equation, finds the root using the bisection method with an error less than 0.001. 

```python
def bisection(f, a, b):
    n = 0
    while (b-a)/2 >= 0.001:
        p=(a+b)/2
        if f(a)*f(p)>0:
            a=p
        else:
            b=p
        n +=1
    print(f"El número de vueltas en el bucle es: {n}")
    print(f"Root found: {p}")
    print(f"\nValor en root: {f(p)}")
def es_real(numero):
    numero_complejo = complex(numero)
    return numero_complejo.imag == 0
```

# Finding the optimal temperature

We are working in a project which aims to grow a new experimental crop. It has been noted that there is a clear correlation between the temperature at which the crop is kept and its final quality (represented as a single value, combination of several metrics). After working for some time in the problem, we have been able to model the relationship between the temperature $T$ of the crop's environment and its quality through the function $f$, given by:

$f(T)=\max(0, -T^4+10T^3-32T^2+38T+54)$

Which would be the optimal temperature for maximizing our crop's quality?

```python
# Code
x = sp.Symbol('x')
y = -x**4+10*x**3-32*x**2+38*x+54
fDerivada = sp.diff(y,x)
print(f"La derivada de la función 1 es: {fDerivada}")
critical_points = sp.solve(fDerivada, x)
print(f"Critical points (puntos en los que la derivada es 0):")
for criticalPoint in critical_points:
    print(f"Derivada en {criticalPoint.evalf():.2f}: {fDerivada.subs(x,criticalPoint).evalf():.2f}")
valoresQueTomanCriticalPoints = [y.subs(x, point) for point in critical_points]

# Find the maximum quality and the corresponding temperature
valorMaximaCalidad = max(valoresQueTomanCriticalPoints)
indiceMaximo = np.argmax(valoresQueTomanCriticalPoints)
temperaturaMaxima = critical_points[indiceMaximo].evalf()

print(f"El punto en el que está el máximo es: {temperaturaMaxima:.2f}")
print(f"El valor del máximo es: {valorMaximaCalidad.evalf():.2f}")
```
